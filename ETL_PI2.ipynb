```python
# Cargar el DataFrame desde el archivo CSV
df = pd.read_csv("Distribución de los accesos totales nacionales a Internet fijo por velocidad de bajada (último trimestre disponible).csv")

# Eliminar la columna "Unnamed: 11"
df = df.drop("Unnamed: 11", axis=1)
# convertimos todas las columnas a str para poder aplicar el replace y eliminar el "." que es un separador de miles
df=df.astype(str)
df = df.applymap(lambda x: x.replace('.', '') if isinstance(x, str) else x)


# se procede a convertir las columnas necesarias a enteros
df[["Trimestre",'Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps',
       'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps',
       'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps',
       'OTROS', 'Total']] = df[["Trimestre",'Hasta 512 kbps', 'Entre 512 Kbps y 1 Mbps',
       'Entre 1 Mbps y 6 Mbps', 'Entre 6 Mbps y 10 Mbps',
       'Entre 10 Mbps y 20 Mbps', 'Entre 20 Mbps y 30 Mbps', 'Más de 30 Mbps',
       'OTROS', 'Total']].astype(int)

# la columna Año se convierte a formato date y solo se mantiene el año
df["Año"] = pd.to_datetime(df["Año"]).dt.strftime('%Y')
df.to_excel('Distribución de los accesos totales nacionales a Internet fijo por velocidad de bajada (último trimestre disponible)modificado.xlsx', index=False)
df.head()


```

```python
df1=pd.read_csv("Ingresos trimestrales de los operadores por el servicio de Internet fijo.csv")
df1=df1[['Año', 'Trimestre', 'Ingresos (miles de pesos)']]

#se realizo la busqueda historica de la cotizacion del dolar por el año y se realiza la conversion
df1.loc[df1['Año'] == 2022, 'Cotizacion dolar'] = 175.39
df1.loc[df1['Año'] == 2021, 'Cotizacion dolar'] = 101.96
df1.loc[df1['Año'] == 2020, 'Cotizacion dolar'] = 83.62
df1.loc[df1['Año'] == 2019, 'Cotizacion dolar'] = 58.43
df1.loc[df1['Año'] == 2018, 'Cotizacion dolar'] = 36.96
df1.loc[df1['Año'] == 2017, 'Cotizacion dolar'] = 18.48
df1.loc[df1['Año'] == 2016, 'Cotizacion dolar'] = 15.82
df1.loc[df1['Año'] == 2015, 'Cotizacion dolar'] = 12.90
df1.loc[df1['Año'] == 2014, 'Cotizacion dolar'] = 8.48

# se convierte la columna a str para realizar la aliminacion de los puntos como separadores de miles luego se convierte en entero
df1['Ingresos (miles de pesos)'] = df1["Ingresos (miles de pesos)"].astype(str)
df1["Ingresos (miles de pesos)"] = df1["Ingresos (miles de pesos)"].apply(lambda x: x.replace('.', '') if isinstance(x, str) else x)
df1['Ingresos (miles de pesos)'] = df1["Ingresos (miles de pesos)"].astype(int)


# se realiza la conversion de pesos argentinos a dolares segun el tipo de cambio historico anual
df1["ingresos (miles de dolares)"]= round(df1['Ingresos (miles de pesos)']/ df1["Cotizacion dolar"],2)    

df1.to_excel("Ingresos trimestrales de los operadores por el servicio de Internet fijo.xlsx", index=False)

df1.head()
```
